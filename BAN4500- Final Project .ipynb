### 5.2.1 RELATIONSHIP BETWEEN NUMBER OF PUBLISHED LECTURES AND NUMBER OF SUBSCRIBERS

A bar graph is used to see the number of subscribers against the number of published lectures for the courses.

```python
df_clients_less_published=df.groupby('num_published_lectures', as_index=False)['num_subscribers'].sum()
df_clients_less_published=df_clients_less_published.head(10)
print(df_clients_less_published)
value = df_clients_less_published.loc[df_clients_less_published['num_published_lectures']==df_clients_less_published['num_published_lectures'].min(), 'num_subscribers'].iloc[0]
print('\n Number of clients that chose course with less number of publised lectures : ' + str(value))
sns.set_theme(style="whitegrid")
#colors=('#48D1CC','#48D1CC','#48D1CC','#48D1CC','#48D1CC','#48D1CC','#48D1CC','#48D1CC','#0000CD')
ax = sns.barplot(x="num_published_lectures", y="num_subscribers", data=df_clients_less_published)
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')
```

### 5.2.5 FINDING THE RELATIONSHIP BETWEEN VARIABLES 

#### <font color='red'>QUESTION : Does number of subscribers depend on the rating, reviews, price and discount of the course?</font>

We will use multiple regression (used to find the dependency of one variable on many other variables) to find whether the number of subscribers for a course depends on the rating and reviews of the course, and on the price and discount provided for the course. 

Let us do hypothesis testing as mentioned in section 3.2.1 : 

<font color='red'>The null hypothesis: H0:</font> Number of subscribers does depend on ratings, reviews, price of course and discount provided.

<font color='red'>The alternative hypothesis: Ha:</font> Number of subscribers does not depend on ratings, reviews, price of course and discount provided.

<font color='red'> EVIDENCE : </font> https://elearningindustry.com/ratings-and-reviews-in-elearning-8-reasons-important

Code inferred from: STAT4600 homework 

```python
#creating a subset containing our independent variables and dependent variable
Sub = df1.filter(items=['num_subscribers','avg_rating_recent','num_reviews','price_detail__amount','discount_price'])
Sub.head()
```

```python
#using scatter plot to visualize the relationship between each variable in the regression model
Scatter_matrix=pd.plotting.scatter_matrix(Sub, figsize=(15, 15), marker='o',
                                 hist_kwds={'bins': 20}, s=60, alpha=.8)
```

The above is a scatter plot matrix that simply shows the relationship between the variables. 

We will focus more on the next step:

```python
#renaming the columns for easy use
Sub.rename(columns={"num_subscribers": "subs", 
                    'avg_rating_recent': 'rate',
                    'num_reviews':'revs',
                    'price_detail__amount': 'price',
                    'discount_price':'discount'}, inplace=True)
Sub.head()
```

```python
#mulitple regression
import statsmodels.formula.api as smf
model_fit = smf.ols('subs ~ rate + revs + price + discount', data=Sub).fit()
model_fit.summary()
```

#### <font color='red'>RESULT :</font>

<font color='red'>R-squared : </font>
The R-squared value depicts the extent to which any variation in the dependent variable (a variable whose value depends on some other factors) is explained by the independent variables (a variable whose value does not depend on any other factors). Here, the dependent variable is number of subscribers and the independent variables are number of reviews, price of the course and discount offered. The value of R-squared is 0.631 (63.1%) which is very high and thus we can say this model is very good and the dependent variable can be described well by the independent variables. 

<font color='red'>F static : </font>
As we can see, the Probability F statistic is 0. This indicates that the model is a good fit. 

<font color='red'>P value : </font>
The P-value (The p-value for each term tests the null hypothesis that the coefficient is equal to zero, i.e.,no effect A low p-value indicates that you can reject the null hypothesis) is 0 for two variables but is 0.116 for price. SInce it is less than 0.1 for rate, reviews and discount, this shows that these variables are significant. But the P-value for price is greater than 0.1 and thus it is not significant for explaining number of subscribers.

<font color='red'>Coefficients : </font>
Finally, we can see that the reviews and discount has positive correlation with number of subscribers as their coefficients are positive. This means that for every increase in numberof reviews, the number of subscribers increases by around 4-5 people. For every increase in discount offered for a course, the number of subscribers increase by 0.36 people. However, the price has negetive relationship with number of subscribers. This means that as price increases by one INR, the number of subscribers decrease by 0.1 people.

#### Thus we can conclude that the number of subscribers for a course does depend on ratings, reviews and discount offered.
